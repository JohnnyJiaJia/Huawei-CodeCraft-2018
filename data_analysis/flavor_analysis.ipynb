```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
path = '../data/test/data_2016_1.txt'
data = pd.read_csv(path,delimiter='\t',header=None,names=['id','flavor','datetime'])
data.head()
```

```python
string = '2016-01-01 00:49:08'
date = datetime.strptime(string, "%Y-%m-%d %H:%M:%S")
print(date)
print(date.strftime("%Y-%m-%d"))
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
#data['datetime'].apply(lambda string : datetime.strptime(string, "%Y-%m-%d %H:%M:%S").strftime("%Y-%m-%d"))

data['day'] = data['datetime'].dt.day
data
```

```python
day_all = data['day'].drop_duplicates()  
day_all.reset_index(drop=True, inplace=True)
day_all
```

```python
flavor_data = data.groupby(['flavor','day'])['datetime'].count().reset_index()
flavor_data.rename(columns={'datetime': 'count'}, inplace=True)

flavor_data
```

```python
flavor1_data = flavor_data[flavor_data['flavor']=='flavor1']
flavor1_data
```
